# Exploratory Data Analysis

## How Many People are Rent a Bike in 2011 and 2022 for Each Month?

### Grouping Data by Year

```python
Year_2011=data_day[data_day.Year==2011]
Year_2012=data_day[data_day.Year==2012]
```

### Data Year of 2011 by Month

```python
Year_2011 = Year_2011.resample(rule='M', on='Date').agg({
    "Count_Casual": "sum",
    "Count_Registered": "sum",
    "Count_Total": "sum"
})
Year_2011.index = Year_2011.index.strftime('%B')

Year_2011['Year']='2011'

Year_2011 = Year_2011.reset_index()

Year_2011
```

### Data Year of 2012 by Month

```python
Year_2012 = Year_2012.resample(rule='M', on='Date').agg({
    "Count_Casual": "sum",
    "Count_Registered": "sum",
    "Count_Total": "sum"
})
Year_2012.index = Year_2012.index.strftime('%B')

Year_2012['Year']='2012'

Year_2012 = Year_2012.reset_index()

Year_2012
```

### Concat Data Years of 2011 and Years of 2012

```python
By_Year=pd.concat([Year_2011,Year_2012])
```

```python
By_Year=By_Year.loc[:,['Date','Year','Count_Casual','Count_Registered','Count_Total']]
By_Year
```

### Save Data

```python
By_Year.to_csv('Data by Year.csv')
```

## How Many People are Rent a Bike for Each Season?

### Grouping Data by Season

```python
By_Season=data_day.groupby(by="Season").agg({
    "Count_Casual": "mean",
    "Count_Registered": "mean",
    "Count_Total": "mean"
}).sort_values(by="Count_Total",ascending=False)
By_Season=By_Season.reset_index()
By_Season
```

### Save Data

```python
By_Season.to_csv('Data by Season.csv')
```

## Visualization and Explanatory Analysis

### How Many People are Rent a Bike in 2011 and 2022 for Each Month?

```python
plt.figure(figsize=(13, 5))
fig1= plt.plot(Year_2011["Date"], Year_2011["Count_Total"],label ='2011', marker='o', linewidth=2, color="orange")
fig2=plt.plot(Year_2012["Date"], Year_2012["Count_Total"],label='2012',marker='o', linewidth=2, color="blue")
plt.title("Bike Rental by Year and Month", loc="center", fontsize=18, weight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylabel("Month")
plt.ylabel("Count of Bike Rental")
plt.legend()
plt.show()
```

Total penyewaan sepeda pada tahun 2012 lebih banyak dibandingkan dengan total penyewaan sepeda pada tahun 2011. Namun apabila dilihat berdasarkan bulan, tren total penyewaan di tahun 2011 dan tahun 2012 hampir sama, yakni dominan naik pada awal tahun saampai pertengahan tahun. Lalu, perlahan turun dari pertengahan tahun sampai akhir tahun. Untuk tahun 2011, pada pertengahan tahun sampai akhir tahun trennya turun. Namun untuk tahun 2012, pada pertengahan tahun sampai akhir tahun trennya naik.

Total penyewaan sepeda terbanyak di tahun 2012 terjadi di bulan September dengan banyak penyewaan sebesar 21.8573.Sedangkan Total penyewaan sepeda terbanyak di tahun 2011 terjadi di bulan Juni dengan banyak penyewaan sebesar 14.3512.

Hal ini menandakan terdapat tren naik penyewaan sepeda di tahun berikutnya karena penyewaan terbanyak pada tahun 2011 dan tahun 2012 cukup berbeda.

Untuk kesediaan unit sepeda juga dapat diperbanyak di awal tahun untuk memenuhi kebutuhan konsumen dan mengoptimalkan biaya operasi perusahaan. 

Dapat pula memberikan promo akhir tahun untuk menstimulasi ketertarikan masyarakat dalam menyewa sepeda.

### How Many People are Rent a Bike for Each Season?

```python
plt.figure(figsize=(10, 5))
sns.barplot(
    y=By_Season["Count_Total"],
    x=By_Season["Season"],
    data=By_Season.sort_values(by="Count_Total", ascending=False),
    color = 'cyan'
)
plt.title("Bike Rental by Season", loc="center", fontsize=15,weight='bold')
plt.ylabel('Average of Bike Rental')
plt.xlabel('Season')
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

Rata-rata penyewaan sepeda terbanyak terjadi pada saat musim dingin yaitu sebesar 5.644 penyewaan, sedangkan pada saat Musim Semi, penyewaan sepeda memiliki rata-rata terendah yaitu sebesar 2604.

Pada musim dingin, perusahaan dapat menambah ketersediaan sepeda untuk memenuhi kebutuhan konsumen.
